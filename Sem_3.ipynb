## 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
stat['Cpu_Company'] = 'Company'
stat.head()
```

```python
stat['Cpu_Company'] = stat['Cpu']
stat.head()
```

```python
def separator (value):
    new_list = []
    for i in value:
        while  i != ' ':
            new_list.append(i)
            break 
        if i == ' ':
            break
    result = ""
    for g in new_list:
        result = result + g
    return result
```

```python
stat['Cpu_Company'] = stat['Cpu_Company'].apply(separator)
```

```python
stat.head()
```

## 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
stat['Memory_Amount'] = 0
```

```python
## stat['Memory_Amount'] = stat['Memory'].apply(separator)
```

```python
def convert_to_gb(x):
    memory = x.split(' ')[0]
    if memory.endswith('GB'):
        res = memory.replace('GB', '')
    elif memory.endswith('TB'):
        res = float(memory.replace ('TB', ''))*1024
    return int(res)
```

```python
stat['Memory_Amount'] = stat['Memory'].apply(convert_to_gb)
```

```python
stat.head()
```

## 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
stat['Memory_Type'] = 0
stat.head()
```

```python
# def separator_2 (value):
#     new_list = []
#     for i in value:
#         while  i != ' ':
#             continue
#         else:
#             new_list.append(i)
#         break
#         if i == ' ':
#             break
#     result = ""
#     for g in new_list:
#         result = result + g
#     return result
```

```python
def convert_to_type(x):
    memory_type = x.replace(' ','').split('B')
    return memory_type[1]
```

```python
stat['Memory_Type'] = stat['Memory'].apply(convert_to_type)
```

```python
stat.head()
```
