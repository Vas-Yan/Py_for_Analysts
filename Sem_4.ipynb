## 3.2 Постройте график в относительных величинах

```python
plt.figure(figsize=(10,6))
plt.bar(ct['Company'],ct['Intel'] )
plt.bar(ct['Company'], ct['AMD'], bottom = ct['Intel'])
plt.title('Laptops by Manufacturers and CPU Producers')
plt.xlabel('Laptop Manufacturers')
plt.ylabel('CPU Manufacturers')
plt.xticks(rotation = 45)
plt.legend(['Intel', 'AMD']);
```

## 3.1 Постройте график в абсолютных величинах

```python
ct2 = pd.crosstab(lp['Company'], lp['Cpu_Company']).reset_index().sort_values(by = 'Intel', ascending=False)
ct2.drop(columns='Samsung',inplace= True)
ct2
```

```python
plt.figure(figsize=(10,4))
ticks = np.arange(len(ct2['Intel']))
offset = 0.2
w = 0.4
plt.bar(ticks - offset, ct2['Intel'], width = w)
plt.bar(ticks + offset, ct2['AMD'], width= w)
plt.xlabel('LP Manufacture')
plt.ylabel ('Amount')
plt.legend(['Intel', 'AMD'])
plt.xticks(ticks, ct2['Company'], rotation = 45);
```

# Задание 5
## Постройте матрицу корреляции

```python
corr_matrix = lp.corr()
corr_matrix = np.round(corr_matrix,1)
corr_matrix[np.abs(corr_matrix)<0.01]=0
corr_matrix
```

```python
plt.figure(figsize=(5,5))
sns.set(font_scale = 0.8)
sns.heatmap(corr_matrix,annot=True,linewidths=0.5,cmap = 'coolwarm')
plt.title('Correlation Matrix');
```

### На основе анализа данных матрицы корреляций прослеживается значимая прямая корреляция между весом ноутбука и диагональю экрана. Остальные парамтеры обладают незначительной корреляцией
